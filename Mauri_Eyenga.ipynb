# Q2: What are the institutions that have been more affected in terms of number of cancelled grants. How does this compare to the others?

```python
terminations_per_org = (
    terminations["org_name"]
    .value_counts()
    .reset_index()
    .rename(columns={"count": "terminations"})
)

terminations_per_org
```

```python
import pandas as pd

# Sort descending by terminations
terminations_per_org_sorted = terminations_per_org.sort_values(
    "terminations", ascending=False
)

# Keep top 5
top5 = terminations_per_org.head(5)

# Aggregate the rest
rest = terminations_per_org_sorted.iloc[5:]
rest_mean = pd.DataFrame(
    {"org_name": ["Avg. Other"], "terminations": [rest["terminations"].mean()]}
)

# Combine top5 and the "Other" row
final_df = pd.concat([top5, rest_mean], ignore_index=True)

final_df
```

```python
import altair as alt

# Horizontal bar chart, ordered by number of terminations
bar_chart = (
    alt.Chart(final_df)
    .mark_bar()
    .encode(
        x=alt.X(
            "org_name:N",
            sort=alt.EncodingSortField(
                field="terminations",  # Sort by terminations
                order="descending",
            ),
            title="Organization",
        ),
        y=alt.Y("terminations:Q", title="Number of Terminations"),
        tooltip=["org_name:N", "terminations:Q"],
    )
    .properties(
        width=600, height=400, title="Number of Terminations per Organization"
    )
)

bar_chart
```

```python
import altair as alt
import pandas as pd

# Expand the counts so each termination is a row
long_df = terminations_per_org.loc[
    terminations_per_org.index.repeat(terminations_per_org["terminations"])
].reset_index(drop=True)

# Jitter plot
jitter_chart = (
    alt.Chart(terminations_per_org)
    .transform_calculate(
        # Add random jitter for X
        jitter="sqrt(-2*log(random()))*cos(2*3.14159265359*random())"
    )
    .mark_circle(size=60, opacity=0.5)
    .encode(
        y=alt.Y("terminations:Q", title="Number of Terminations"),
        x=alt.X("jitter:Q", title=None, axis=None),
        tooltip=["org_name:N"],
    )
    .properties(
        width=200,
        height=400,
        title="Terminations per Organization (Jittered Points)",
    )
)

jitter_chart
```
